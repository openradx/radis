```python
report = """
Patient Name: John Doe
Medical Record Number: 1234
Exam: CT Abdomen and Pelvis with Contrast

Clinical History: 45-year-old male with history of abdominal pain and weight loss.
Technique: Axial images of the abdomen and pelvis were obtained post administration of intravenous contrast.

Findings:
- There is a 6 cm mass identified in the lower pole of the left kidney, consistent with a renal cell carcinoma.
- No evidence of metastatic disease identified in the liver, lungs, or bones.
- There are multiple enlarged lymph nodes in the retroperitoneum, likely representing nodal metastases.
- Mild dilation of the pancreatic duct is noted, suggestive of possible pancreatic ductal adenocarcinoma.
- Mild thickening of the sigmoid colon wall, concerning for possible inflammation or neoplastic process.

Impression:
1. Renal cell carcinoma in the left kidney with nodal metastases.
2. Possible pancreatic ductal adenocarcinoma.
3. Mild sigmoid colonic wall thickening, further evaluation recommended.

Recommendation:
Consultation with appropriate specialists for further management and treatment options.
Follow-up imaging and possible biopsy of the kidney mass and pancreatic ductal lesion.
"""
```

```python
import openai
from pydantic import BaseModel

class Output(BaseModel):
    possible_cancer: bool
    possible_metastasis: bool
    cancer_location: str
    metastasis_location: str

client = openai.OpenAI(
    base_url=f"http://0.250.250.254:11434/v1", api_key="ollama"
)

system_prompt = """
You are a medical doctor that should provide structured data of a radiology report.
Only provide data that is really found in the report. Don't make up new data.
Don't hallucinate or provide data that is not supported by the report.

Fields:
possible_cancer: Whether cancer is mentioned in the report.
possible_metastasis: Whether metastasis is mentioned in the report.
cancer_location: Possible cancer locations mentioned in the report (separated by commas).
metastasis_location: All locations suspected of metastasis in the report (separated by commas).
"""

response = client.beta.chat.completions.parse(
  model="qwen2.5:14b",
  messages=[
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": report},
  ],
  response_format=Output,
)
response.choices[0].message.parsed
```
